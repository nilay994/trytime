## Load raw Accelerometer and Gyro readings from AlphaPilot

```python
import os
import datetime
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import sys
```

```python
# acc file
acc_flag = False
acc_file = 'accel_data.csv'
if os.path.isfile(acc_file):
    if os.stat(acc_file).st_size != 0:
        acc = np.genfromtxt(acc_file, delimiter=',', skip_footer=1)
        acc_flag = True

        t_acc = acc[:,0] + 10**(-9) * acc[:,1]
        t_acc = t_acc - t_acc[0]
        acc_data = acc[:,2:-1]
```

```python
# gyr file
gyr_flag = False
gyr_file = 'gyro_data.csv'
if os.path.isfile(gyr_file):
    if os.stat(gyr_file).st_size != 0:
        gyr = np.genfromtxt(gyr_file, delimiter=',', skip_footer=1)
        gyr_flag = True

        t_gyr = gyr[:,0] + 10**(-9) * gyr[:,1]
        t_gyr = t_gyr - t_gyr[0]
        gyr_data = gyr[:,2:-1]
```

## Plot raw sensor readings

```python
if gyr_flag:
    plt.figure()
    plt.plot(t_gyr, gyr_data)
    plt.xlabel('t')
    plt.suptitle('Gyro')
    plt.grid(True)
    plt.show()
```

```python
if acc_flag:
    plt.figure()
    plt.plot(t_acc, acc_data)
    plt.xlabel('t')
    plt.suptitle('accel')
    plt.grid(True)
    plt.show()
```

0 & \cos (\hat{\phi}) & -\sin (\hat{\phi})
\end{array}\right]\left[\begin{array}{l}
p \\
q\\\
r
